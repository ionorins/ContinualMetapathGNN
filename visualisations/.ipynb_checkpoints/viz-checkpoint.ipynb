```python
import numpy as np
import matplotlib.pyplot as plt

timeframes = 25
aspects = ('continual', 'online', 'pretrained', 'retrained', 'single')

for aspect in aspects:
    data = np.genfromtxt(f'{timeframes}{aspect}future.csv', delimiter=',')
    data = data[:,1]
    plt.plot(data, label=aspect)

plt.legend(loc='upper right')
```

```python
def speedup(theta):
    ret = sum(range(1, 24))
    cont = 0
    for i in range(1, 24):
        cont += i if i < theta else theta
    return ret/cont

x = range(1, 24)
y = [speedup(theta) for theta in x]

plt.plot(x, y)
```
