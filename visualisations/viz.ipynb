```python
def speedup(theta):
    ret = sum(range(1, 24))
    cont = 0
    for i in range(1, 24):
        cont += i if i < theta else theta
    return ret/cont

x = range(1, 24)
y = [speedup(theta) for theta in x]

plt.plot(x, y)
```
